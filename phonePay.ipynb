```python
district_txn_agg = District_Txn_Users.groupby("District", as_index=False)[["Transactions"]].sum()
```

```python
district_merged = pd.merge(district_txn_agg, District_Demographics, on="District", how="inner")
```

```python
density_corr = district_merged["Density"].corr(district_merged["Transactions"])
plt.figure(figsize=(10,6))
sns.scatterplot(x="Density", y="Transactions", data=district_merged)
plt.title(f"Population Density vs Transaction Volume (r = {density_corr:.2f})")
plt.xlabel("Population Density")
plt.ylabel("Total Transactions")
plt.grid()
plt.show()
```
